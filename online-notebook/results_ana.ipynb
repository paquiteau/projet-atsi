```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import collections
data = np.load('/home/pac/gits/phd/mri-online/online-notebook/grid_search_2021-05-21_14:36:08.npy',allow_pickle=True).item()

results_metrics = collections.defaultdict(dict)
for simul, res  in data.items():
    print(simul)
    for metric_name, metric_data in res.items():
        results_metrics[metric_name][simul] = metric_data
print(results_metrics.keys())
```

```python
metric_name='cost'
simul = results_metrics[metric_name]
for idx, (simul_name, simul_result) in enumerate(simul.items()):
    if idx % 10 == 0:
        if idx > 0:
            plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
        plt.figure()
        plt.title(f'{metric_name}')
    plt.semilogy(simul_result['index'],simul_result['values'],label=simul_name)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

```
